```python
import matplotlib.pyplot as plt

# Create data for multiple lines
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 1, 8, 27, 64, 125]
y3 = [0, -1, -4, -9, -16, -25]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot multiple lines
ax.plot(x, y1, label='y = x^2', marker='o')
ax.plot(x, y2, label='y = x^3', marker='s')
ax.plot(x, y3, label='y = -x^2', marker='^')

# Set axis labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

# Add a title
ax.set_title('Multi-Line Plot Example')

# Add a legend
ax.legend()

# Add grid lines
ax.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import requests

df = pd.DataFrame()

for i in range(1, 429):
    response = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=eaa36bac944f1b3d3f9e954931338d77&page=' + str(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'original_language', 'original_title', 'title', 'overview', 'popularity', 'release_date', 'vote_average', 'vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)

```
